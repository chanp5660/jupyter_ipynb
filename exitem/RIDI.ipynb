# RIDI

> 꼭 같이 일 해보고 싶은 회사이다.  

> 처음 리디를 만났을 때는 2017년 대학교 2학년 보조강사 알바를 할때 공무원 선생님이 사용하지는 이북리더기를 봤다. 노트북 산다고 한참 기기란 기기는 가리지 않고 좋아했을 때라 책만 읽을 용도인 기기 참을 수 없었다. 하지만 가격이... 생각보다 비쌌다. 물론 기기를 사고 E-book을 산다면 총량은 싸겠지만 당장 E-book으로 나와있는 유명한 책들도 많지 않은 상태였다. 그 후로 잊고 지내다가 어느날 보이는 리디 웹툰이라는 글을 보게 되었고 웹툰까지 하는구나 하고 넘겼는데 최근에 와서는 유니콘 기업으로 선정되었다는 말까지 들렸다. 내가 거물과 같이 크지 못했던 아쉬움이 남았고 내 머리에 기억이 박힐만한 충격이었고 앞으로의 행보가 기대가 되는 회사이다.  


## RIDI 회사 기본정보 

[공식 사이트](https://ridicorp.com/about/) 및 [위키피디아 참조](https://ko.wikipedia.org/wiki/리디주식회사)


- 리디주식회사(영어: RIDI Corporation)는 2008년 3월 설립된 콘텐츠 플랫폼 기업이다.
- 2009년 리디북스 서비스를 시작했으며, **전자책 1위**를 발판삼아 웹툰, 웹소설, 애니메이션 등으로 사업을 확장하고 있다. 
    - <span style="color:blue">전용 기기를 같이 낼 생각을 한 것은 굉장한 시도라고 생각한다. 태블릿이랑 스마트폰이 많은 세상이라 그런지 나는 생각지도 못한 부분이었다.</span>
- 2022년에는 대표 서비스인 ‘리디북스’의 명칭을 ‘리디’로 변경하며 콘텐츠 플랫폼으로 도약을 선언했다. 리디 외에도 다양한 서비스를 운영한다. 
- 대표적으로는 2020년에 글로벌 웹툰 구독 서비스 만타를 북미에 출시했다.
    - <span style="color:blue">이 부분은 정말 존경하고 싶다. 어릴 때부터 웹툰을 본 나는 한글로만 나온 웹툰들이 아쉬웠고 충분히 외국에서도 먹힐수도 있을거라는 생각이 들었고 상품성이 커지면 좋은 결과물로 또 작가님들과 볼 수 있는 기회가 생길 것 같아 정말 감사한 부분이다.</span>
- 실적도 꾸준히 성장 중이다. 2010년 4분기부터 2021년 3분기까지 44분기 연속 성장했다고 발표했고, 2021년 연매출은 2037억원을 기록하며 전년 대비 30.9% 성장했다.
    - <span style="color:blue">성장한 시기에 세계적으로 국가적으로도 어려운 시기가 있었지만 계속 성장을 한 RIDI라는 회사는 정말 존경스럽다.! 아이디어가 좋은 회사인 것 같다.</span>
- 2022년 2월에는 싱가포르투자청(GIC) 등으로부터 1200억원 규모의 투자를 유치했고, 그 과정에서 1조 6천억원의 기업가치를 인정 받아 국내 콘텐츠 플랫폼 스타트업으로는 처음으로 유니콘 기업에 등극했다.
- 사원수 : 489
    - <span style="color:blue">입사자 수, 퇴사자 수도 꾸준히 감소 하는 것을 보면 회사가 보다 안정적으로 되고 있는 것 같다.</span>

## 서비스

### 역사
- 2009년 11월 리디북스 출시하며 전자책 사업을 시작했다. 당시에는 전자책 자체가 생소했기에 출판사들과 협업도 쉽지 않았지만, 꾸준히 전자책의 강점을 알리고 소프트웨어 경쟁력을 높여 업계 1위에 올랐다.
- 2018년부터는 웹소설과 웹툰, 애니메이션 등으로 콘텐츠 사업을 확장하기 시작했다. 현재는 웹소설을 웹툰화하는 노블코믹스 제작과 동시에, 웹툰 공모전 등을 통한 작품 발굴도 추진하고 있다.
- 2020년 11월에는 북미 시장에 웹툰 구독 서비스 만타(Manta)를 출시했다. 웹툰 업계에서는 처음으로 구독 모델을 적용했고, 리디의 독점작 등 다양한 콘텐츠를 서비스하고 있다. 론칭 이후 가파른 성장세를 보이며 출시 1년여만에 미국 플레이스토어 만화 앱 부문에서 북미·유럽·아시아 지역 16개국 1위에 오르며 영향력을 키워가고 있다.
