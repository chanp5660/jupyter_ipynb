# 웨인힐스벤처스

> 내가 상상만 했던 것이 현실로 진행한 회사

> 누구나 책을 읽으면서 간략하게 요약해줬으면 좋겠다라는 생각을 한다. 최근에 chatGPT 가 나오기전에 몰랐던 사람도 많고 속도가 빠르지 않거나 원하는 정도의 수준으로 나오지 않아 실사용 되는 경우가 적었다. 웨인힐스벤처스는 더 나아가 요약한 부분을 영상으로 요약해주는 것까지 생각하여 굉장한 흥미와 실사용에도 사용할정도로 결과가 좋은 것 같다. NLP를 공부해보아서 기술적인 부분이 얼마나 어려운 구성인가를 알기 때문에 더 잘 보이는 것 같다. 유투브 홍보영상을 보았는데 대부분 영어로 말씀을 하시는데 ㅠ 영어공부를 역시 해야겠다는 생각이 든다. 회사에 수준에는 못 미칠 수 있지만 합격한다면 나도 회사도 같이 발전할 수 있을 것 같다.
