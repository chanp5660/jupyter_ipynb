# 노리스페이스

> 내가 처음에 생각했던 인공지능의 역할을 비전으로 가지고 있다. 

> 사무업무를 자동화 하는 일에 인공지능이 필요하다는 것은 어릴적 부모님을 도와 회계업무, 보고서 작성 등을 하면서 꼭 필요하다고 생각했고 컴퓨터를 배우면서 점점 할 수 있다는 생각이 들었다. 그 일을 하고 있는 것이 노리스페이스라고 생각한다. 하지만 아직 초창기 스타트업인만큼 안정화 되어 있다고 느껴지지는 않는다. 입사자 수와 퇴사가 수가 거의 일치하는 느낌이라 배울 수 있는 것, 이끌 수 있는 것이 많을 수 있지만 그만큼 분야를 가리지 않고 다방면으로 노력해야 할 것 같다.
