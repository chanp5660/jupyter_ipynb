##  풀스택 개발자(1명)
- 담당업무
    - 기상청 프로젝트 개발 및 운영 업무
    - 기상, 해양, 공간 정보 관련 R & D
- 자격요건
    - 학력 : 대졸이상
    - 경력 : 신입
    - Python 기본적인 문법을 알고 있는 분
    - 리눅스, SQL, javascript 기본 지식이 있으신 분
- 우대사항
    - 평소 소프트웨어 개발에 관심있고 다양한 경험을 도전하신 분
    - 컴퓨터 공학 및 관련 학과 전공자
- 업무환경
    - 공공기관 클라이언트를 상대로 과학자료의 가공 및 표출을 주 업무로 합니다.
    - 회사 정책으로 파견근무를 지양하고 본사 근무를 원칙으로 합니다.(단, 일반적인 수준의 외근과 출장은 있습니다.)
- 기술사항
    - 취급분야가 다양하지만 각각의 난이도는 낮은 편 입니다.
    - 리눅스쉡 활용 능력은 필수적입니다.
    - 수학적인 기능 구현 시 대부분 규격이 제시되기 때문에 관련 능력은 중요하지 않습니다.
    - openlayers, cesium 같은 GIS 플랫폼을 활용하는 경우도 있습니다.
    - python 으로 ascii는 물론 netcdf, grid 등 다양한 포맷 데이터를 취급합니다.
    - ascii 파싱으로 가능하다면 쉽게 적응 할 수 있는 수준입니다.
    - DB는 기초적인 CRUD 쿼리 작성이 가능하면 됩니다.
    - 정해진 서버 환경으로 인해 웹 서버사이드에 php를 자주 적용하지만 그 외에는 제약이 적어서 원하는 기술을 적용하는 자기 계발적 업무가 가능합니다.
