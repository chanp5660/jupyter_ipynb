# 정보전략계획(ISP)

ISP는 Information Strategic Plan의 약자이며 기업이 수립한 중장기 경영 전략과 계획을 토대로 사업 전개에 필요한 총체적인 정보 체계를 제시하고 향후 단위 또는 통합 정보 체계의 개발을 계획하고 통제함으로써 경영 요구에 의한 정보 기술 체계를 구축하는 과정으로 정의할 수 있습니다.

또한, 조직이나 기업이 정보와 기술을 최대한 활용하여 비즈니스 목표를 달성하기 위해 수립하는 계획입니다. ISP는 조직의 비즈니스 목표와 정보 기술(IT) 전략 간의 일치를 보장하며, 정보 시스템과 기술을 효과적으로 관리하고 활용할 수 있는 로드맵을 제시합니다. 장기적인 비전과 중장기적인 목표를 가지며, 조직 내의 다양한 부문과 관련 이해 관계자들 간의 협력을 강화하는데 중요한 역할을 합니다.

활용 : ISP를 수립함으로써 조직은 비즈니스 목표를 더 효과적으로 달성하고 경쟁 우위를 확보할 수 있게 됩니다. ISP는 비즈니스 리더쉽, 정보 기술 전문가 및 기타 이해 관계자들 간의 협력을 필요로 하며, 기술 도입과 조직 변화를 조율하는데 도움을 줍니다.

## 단계

### 비즈니스 목표 및 요구사항 분석

조직의 비즈니스 목표와 전략을 분석하고, 정보 기술의 역할과 기대되는 기능을 식별합니다. 비즈니스 환경 변화에 따른 요구사항도 파악합니다.  

### 기술 및 인프라 평가

현재의 기술 및 정보 시스템 인프라를 평가하고, 이를 개선하거나 혁신적인 기술 도입이 필요한지 판단합니다.  

### 비즈니스와 IT 전략의 일치성 확보

비즈니스 목표와 IT 전략 간의 일치를 검토하고, IT가 비즈니스 성과를 지원할 수 있는 방법을 찾습니다.

### 목표 및 전략 수집

ISP에서 달성하고자 하는 장기적인 목표와 중장기적인 전략을 수립합니다. 이는 비즈니스 성과 개선, 기술 혁신, 정보 보안, 데이터 관리 등을 포함할 수 있습니다.  

### 프로젝트 및 우선순위 결정  

ISP에 따라 추진해야 할 프로젝트를 결정하고, 이를 우선순위에 따라 계획합니다. 이 단계에서 자원 할당과 일정 계획이 이루어집니다.  

### 조직 내 의사소통 및 협업 강화  

ISP의 성공적인 수행을 위해 조직 내부의 다양한 부문과 이해 관계자들 간의 의사소통과 협력을 강화합니다.  

### 성과 평가 및 조정

ISP의 실행이 진행되는 동안 주기적으로 성과를 평가하고, 필요한 경우 전략을 조정하여 비즈니스 목표 달성에 도움을 줍니다.
