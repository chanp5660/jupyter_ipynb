# 분석과제의 목표가치(KPI)

KPI는 Key Performance Indicator의 약자로 분석과제의 목표값은 조직이나 프로젝트의 성과를 측정하고 평가하는데 사용되는 중요한 지표입니다.   
조직의 목표와 전략에 따라 설정되며, 성과를 정량적으로 측정하여 비즈니스의 성공 여부를 파악하고 조정하는데 도움을 줍니다.  
목표값은 조직의 목표 달성 정도를 평가하는 기준이 되며, 이를 통해 조직은 비즈니스 방향성을 조율하고 개선할 수 있습니다.

KPI 분석과제에서 목표값은 조직의 성과 평가와 개선을 위한 중요한 출발점으로서, 명확하게 설정되어야 하며 조직 내에서 이해되고 수용되는 것이 중요합니다.

- 예시
    - 매출 증가가 조직의 주요 목표라면 '월별 매출액'은 중요한 KPI일 수 있습니다. 이 KPI의 목표값은 매월 설정되며, 실제 매출액과 비교하여 목표값에 도달했는지, 미치지 못했는지를 판단합니다. 이를 기반으로 조직은 매출 증가를 위한 전략을 조율하고 개선해 나갈 수 있습니다.

## 특징

### 정량적인 측정

KPI는 수치로 표현되는 지표로, 조직의 성과를 정량적으로 측정합니다. 이를 통해 주관적인 아닌 객관적인 분석이 가능합니다.  

### 비즈니스 목표와 연결

KPI의 목표값은 조직의 비즈니스 목표와 긴밀한 연관이 있어야 합니다. 즉, 조직의 전략과 성과에 직접적으로 기여하는 내용이어야 합니다.

### 성과 평가와 개선

KPI의 목표값을 통해 조직은 성과를 평가하고 개선할 수 있는 기준을 확보합니다. 성과가 예상치에 미치지 못하면 조치를 취하고 성과를 개선하기 위한 방안을 모색할 수 있습니다. 

### 변경 가능성

KPI의 목표값은 시간에 따라 조직의 우선순위나 전략 변화에 따라 조정될 수 있습니다. 조직의 변화에 따라 목표값을 조정하여 비즈니스에 더 적합한 지표를 유지할 수 있습니다.
