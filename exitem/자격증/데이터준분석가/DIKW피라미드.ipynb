# DIKW 피라미드(Data Information Knowledge Wisdom)

DIKW 피라미드는 정보과학과 지식 관리 분야에서 널리 사용되는 개념적인 모델입니다.  
**데이터, 정보, 지식 및 지혜**라는 네 가지 계층으로 구성되어 있으며, **하위 계층에서 상위 계층으로 가는 방향으로 계층화**되어 있습니다.  
각 계층은 이전 계층의 결과물을 기반으로 하여 더 높은 수준의 추상화와 이해를 제공합니다.  

DIKW 피라미드는 이러한 **계층 구조를 통해 데이터가 지혜로 가는 과정**을 설명합니다.  
데이터는 정보로 가공되며, 정보는 지식으로 변환되고, 지식은 지혜의 토대가 됩니다.   
이 모델은 정보 관리, 지식 전달, 의사 결정, 학습 및 혁신 등 다양한 분야에서 활용되며, 지식의 가치를 최대화하기 위한 전략과 방법론의 개발에도 도움이 됩니다.

## 데이터(Data)

단순히 **사실**이나 기록으로서 **원시적**이고 **구조화되지 않은 형태의 정보**입니다.  
개별적인 사실이나 관찰을 나타내며, 그 자체로는 큰 의미나 가치를 지니지 않습니다.   

예를 들어, 온도 측정 데이터, 고객 주문 데이터, 주식 시장 거래 데이터 등이 데이터의 예시입니다.  

## 정보(Information)

데이터를 **조직화**하고 **해석**하며 **의미 있는 내용**을 제공하는 것입니다.  
데이터의 패턴, 관계, 특성 등을 이해하고 분석함으로써 얻어집니다.  
컨텍스트와 의미를 부여하여 가치를 더합니다.  

예를 들어, 온도 데이터를 바탕으로 특정 지역의 날씨 정보를 제공하거나, 고객 주문 데이터를 분석하여 판매 동향 정보를 도출하는 것이 정보의 예시입니다. 

## 지식(Knowledge)

**정보를 이해**하고 **활용**할 수 있는 **인식과 깨달음**을 말합니다.  
개별 정보들의 상호 연결, 패턴 인식, 근거에 기반한 추론, 경험 등을 통해 형성됩니다.  
문제 해결, 의사 결정, 예측 등을 위해 활용 될 수 있습니다.  

예를 들어, 날씨 정보를 바탕으로 특정 지역에서 휴가를 가기 좋은 시기를 예측하는 것입니다.

## 지혜(Wisdom)

**넓은 시야**와 **깊은 이해**를 바탕으로 형성되며, **지식을 토대로 한 윤리적이고 상황에 맞는 판단과 행동을 포함**합니다.  
비지니스 전략 수립, 사회적 문제 해결, 윤리적인 결정 등에 관련되며, 전문적인 지식과 경험이 요구됩니다.  
지혜는 지식을 최고 수준으로 활용하여 긍정적인 영향력을 가지는 행동을 취함으로써 얻어집니다.  
**지식과 경험의 결합으로 형성**되며, 복잡하고 불확실한 상황에서 효과적인 판단을 내릴 수 있는 능력을 강조합니다.  

비즈니스, 리더십, 정치, 인간관계 등 다양한 영역에서 중요한 역할을 합니다.
