# 빅데이터와 인문학

빅데이터와 데이터 사이언스 분야의 성장과 인문학의 열풍 사이에는 몇 가지 외부 환경적 요인들이 상호작용하여 원인이 되었습니다.  

## 데이터의 폭발적인 증가

현대 사회에서는 디지털 기술과 인터넷의 보급으로 많은 양의 데이터가 생성되고 저장됩니다.  
모바일 기기, 소셜 미디어, 인터넷 활동, 센서 등을 통해 생성되는 데이터 양은 기하급수적으로 증가하고 있습니다.  
이러한 폭발적인 데이터 양은 빅데이터의 개념을 부각시켰고, 이 데이터들을 분석하고 이해하는 데이터 사이언스의 중요성을 높였습니다.  

## 데이터 저장 기술의 발전

빅데이터의 등장과 함꼐 데이터 저장 기술도 발전하였습니다.  
예전에는 대규모 데이터를 저장하고 처리하는 것이 어려웠지만, 현재는 저렴한 비용으로 대용량 데이터를 저장하고 분석할 수 있는 기술이 보편화되었습니다.  
클라우드 컴퓨팅과 분산 시스템의 발전은 데이터 사이언스의 기반을 확립하였고, 이는 빅데이터 분석의 가능성을 더욱 높였습니다.  

## 기술의 발전과 접근성의 증가

데이터 사이언스와 빅데이터 분석에 필요한 도구와 기술들이 발전하면서, 비전공자들에게도 접근성이 증가했습니다.  
예전에는 전문적인 프로그래밍 지식과 통계학적인 지식을 요구하는 분야였지만, 현재는 사용자 친화적인 도구들과 온라인 강의 등이 널리 제공되고 있어 더 많은 사람들이 데이터 사이언스를 배우고 활용할 수 있게 되었습니다.  

## 사회적 요구의 변화

현대 사회에서는 데이터 기반의 의사결정과 문제 해결이 중요시되고 있습니다.  
기업, 정부, 비영리 기관 등 다양한 조직과 분야에서 데이터를 활용하여 전략 수립, 고객 분석, 리스트 관리, 인사 관리 등을 진행하고 있습니다.  
이러한 사회적 요구와 관련된 경쟁력 강화, 효율성 향상 등이 데이터 사이언스와 빅데이터 분야의 성장과 인문학의 열풍을 부추겼습니다.
