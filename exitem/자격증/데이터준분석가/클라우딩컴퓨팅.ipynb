# 클라우딩 컴퓨팅(Cloud Computing)

인터넷을 통해 컴퓨터 리소스를 제공하고 관리하는 기술 및 서비스입니다.  
기존에는 개별적으로 소유하고 관리되는 로컬 컴퓨터 또는 서버를 사용하여 데이터를 저장하고 응용 프로그램을 실행했지만, 클라우딩 컴퓨팅은 인터넷을 통해 필요한 컴퓨팅 리소스에 대한 접근성과 유연성을 제공합니다.  

기업과 개인 모두에게 다양한 이점을 제공합니다.   
비즈니스에 있어서는 비용 절감과 유연성을 통해 경제적인 이점을 얻을 수 있으며, 개발자와 IT 전문가에게는 개발과 운영의 간소화와 효율성 향상을 가져옵니다.   
또한, 사용자는 모바일 기기나 웹 브라우저를 통해 언제 어디서든 필요한 서비스에 접근할 수 있으므로 편의성과 업무 유연성을 확보할 수 있습니다.


---

## 주요 특징

### 온디맨드 서비스(On-Demand), 유연성과 확장성

클라우드 서비스는 필요에 따라 리소스를 증가 또는 축소할 수 있으므로 유연성과 확장성이 뛰어납니다.  
사용자는 필요한 만큼의 컴퓨팅 리소스를 사용할 수 있으며, 규모에 따라 리소스를 조정할 수 있습니다.  
이는 사용자의 요구에 따라 유연하게 서버, 스토리지, 대역폭 등의 리소스를 조정할 수 있는 능력을 갖추었음을 의미합니다. 

### 네트워크 접근성

클라우드 서비스는 인터넷을 통해 언제 어디서든 접근할 수 있습니다.  
사용자는 클라우드에 저장된 데이터와 애플리케이션에 대한 접근성을 인터넷을 통해 얻을 수 있습니다. 

### 공유 리소스(협업과 유연한 작업환경)

여러 사용자가 동일한 물리적 리소스를 공유함으로써 비용과 자원 사용률을 최적화합니다.  
이는 여러 사용자가 동시에 클라우드에서 서비스를 이용하며, 필요에 따라 자원을 할당받을 수 있음을 의미합니다. 

여러 사용자가 동시에 작업하고 데이터 및 문서를 공유할 수 있습니다.  
협업 도구, 파일 공유, 팀 프로젝트 관리 등을 제공하여 효율적인 작업 환경을 조성할 수 있습니다.

### 유지보수 및 관리 간소화

클라우드 서비스 제공업체는 시스템의 유지보수, 업그레이드, 보안 등을 관리하므로 사용자는 이에 대한 부담과 관리 비용을 줄일 수 있습니다.  
사용자는 클라우드 서비스에 대한 기술적인 측면을 걱정하지 않고 핵심 비즈니스에 집중할 수 있습니다.  

클라우드 서비스 제공업체는 데이터 보안에 대한 향상된 기술과 보안 매커니즘을 갖추고 있습니다.  
데이터 암호화, 접근 제어, 모니터링, 백업 및 복구 등의 보안 기능을 제공하여 데이터의 기밀성과 무결성을 보호합니다.  

그러나 클라우드 사용자는 데이터 보안을 위해 적절한 접근 권한 관리와 추가적인 보안 조치를 적용해야 합니다. 

### 비용 절감

초기 투자 비용을 줄일 수 있습니다.  
사용자는 인프라를 구매하고 유지보수하는 대신 사용한 만큼 비용을 지불하게 됩니다.  
또한, 클라우드 서비스는 사용량에 따라 가격이 유동적으로 책정되므로 비용을 최적화할 수 있습니다.  

### 고가용성과 신뢰성

여러 가용 영역에 데이터 및 애플리케이션을 분산시키고 복제함으로써 고가용성과 신뢰성을 제공합니다.  
이는 단일 서버 또는 인프라 장애로부터의 회복력을 향상시키고, 시스템의 가동 시간을 최대화할 수 있습니다.

## 서비스 모델

### 인프라스트럭처 as a Service (IaaS)

가상화된 컴퓨팅 리소스를 제공하는 서비스입니다.  
사용자는 가상 서버, 스토리지, 네트워킹 등의 인프라를 필요에 따라 사용할 수 있습니다.  
이 모델은 자체적으로 하드웨어를 관리하고자 하는 기업에게 유용하며, 서버나 네트워크 인프라 등을 클라우드로 외부에 위탁할 수 있습니다. 

### 플랫폼 as a Service (Paas)

애플리케이션 개발과 운영을 위한 플랫폼을 제공하는 서비스입니다.  
개발자는 클라우드 플랫폼에서 애플리케이션을 개발, 테스트, 배포 및 관리할 수 있습니다. 
개발자가 인프라 관리보다 애플리케이션 개발에 더 집중할 수 있도록 지원합니다. 

### 소프트웨어 as a Service (SaaS)

애플리케이션을 인터넷을 통해 제공하는 서비스입니다.  
사용자는 클라우드 상에서 제공되는 소프트웨어를 웹 브라우저나 모바일 앱을 통해 접근하여 사용할 수 있습니다.  
이 모델은 일반 사용자와 기업에게 소프트웨어를 구매하지 않고도 필요한 기능을 이용할 수 있는 편리성을 제공합니다.
