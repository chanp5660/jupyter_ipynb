# 분석 기회 발굴 범위

분석 프로젝트에서 **조사 및 분석할 대상 및 범위를 결정하는 것을 의미**합니다.  
이는 분석 목적과 목표를 달성하기 위해 필요한 데이터와 리소스에 대한 범위를 정하는 과정입니다.  

분석 기회 발굴 범위를 명확히 정의하고 이해함으로써 효율적인 분석을 수행할 수 있으며, 비즈니스 목표 달성에 필요한 통찰력과 결과를 도출할 수 있습니다.

## 비즈니스 목표

분석 기회는 비즈니스 목표와 관련되어야 합니다.  
**어떤 문제를 해결**하고자 하는지, **어떤 비즈니스 결과를 도출**하고자 하는지를 명확히 이해해야 합니다.

## 데이터 

분석에 필요한 **데이터를 식별**해야 합니다.  
어떤 **종류**의 데이터가 필요하고, 어떻게 **수집**되는지, 사용 가능한 데이터 세트의 **범위와 품질**을 평가해야 합니다.

## 분석 범위

분석할 **대상의 범위**를 정해야 합니다.  

예를 들어, 특정 제품, 고객 세그먼트, 지역 또는 기간 등에 초점을 맞출 수 있습니다.  
이는 분석 목표와 관련하여 가장 중요한 영역을 식별하는 것입니다.

## 리소스

분석에 **사용할 수 있는 리소스(인력, 시간, 도구 등)를 고려**해야 합니다.  
사용 가능한 리소스에 따라 분석 범위를 조정해야 할 수도 있습니다.
