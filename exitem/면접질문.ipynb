> 면접 때 질문 받은 부분들을 정리해서 다음 면접 때 해당 질문에 답할 수 있도록 진행합니다.

## 간단한 자기소개 해주세요.
- 네 안녕하세요! 지원자 박찬입니다.
- 통계학과에 진학하여 적은 데이터로 추론, 추측을 하다가 2학년 수업에서 python을 이용한 머신러닝 수업을 접하게 되면서 컴퓨터공학부를 부전공하고 교수님과 인연이 되어 컴퓨터공학부 대학원에 진학하게 됐습니다.
- 석사 중에는 범부처 협업을 하게 되면서 농산물 관련 온라인 소셜 미디어 데이터를 통해 농부들에게 가격변동에 영향을 주는 요인들을 제공해주는 프로젝트를 진행했습니다. 
- 이후 1년 공백기에는 부모님이 산림조합을 설립 하셨는데 컴퓨터를 잘 다루지 못하시고 인력도 구하기 힘든 부분이라서 도와드렸습니다.
- 현재 2월 중순부터 취업준비를 시작하여 이 면접을 보게 된 것입니다.


## 프로젝트에 대해서 자세히 설명 해주세요.
- 네, 우선 프로젝트의 결과부터 말씀드리겠습니다.
- 농부들에게 가격데이터와 온라인 소셜미디어 데이터를 이용하여 정보를 제공하는 것입니다.
- 세부적으로 살펴보면 가격변동 데이터를 맡은 팀에서 가격변동을 제공하고 특정 기간과 품목을 설정하면 오름세와 내림세의 주요 요인과 관련된 SNS 정보들을 제공하는 프로젝트입니다.
- 이를 통해 농부들은 해당 품목에 대해서 어떤 요인들이 발생했는지를 파악하고 조심하여 행동변화를 일으키거나 걱정을 덜어드릴 수 있습니다.
- 기술적으로 살펴보면 데이터 수집은 KAMIS(농산물유통정보) 같은 전문적인 사이트에서 주요 용어를 얻어 네이버, 구글, 트위터에서 api와 웹 스크랩핑(selenium,BeautifulSoup)으로 수집했습니다. pysql을 이용해서 데이터 저장 및 관리 했습니다. 문서 유사도로 비슷한 문서들을 모아놓고 토픽 모델링을 사용하여 데이터에서 레시피, 영양성분 등의 가격변동에 영향을 주는 주요 용어를 파악했습니다.
