# 데이터의 유형(정성,정량)


**정성적** 데이터와 **정량적** 데이터는 데이터의 특성에 따라 분류된 두 가지 유형의 데이터입니다.  
데이터의 특성과 활용 방안에 따라 차이가 있으며, 분석 목적에 맞게 적절한 데이터를 선택하여 활용하는 것이 중요합니다.

## 정성적 데이터

- 주로 **질적** 데이터로 불리며, 수치화되지 않은 데이터를 의미합니다.  
- 숫자나 측정 단위가 없는 비수량적인 데이터를 의미하며 ,대개 언어, 이미지, 비디오, 소리 등의 형태로 표현됩니다.
- 예를 들어, 고객 만족도 조사에서 고객의 만족도를 '매우 만족', '만족', '보통', '불만족', '매우 불만족'과 같은 카테고리로 평가하는 경우가 있습니다.
- 이러한 데이터는 주로 감성적인 분석이 필요하거나, 인과관계나 통계적 분석이 어려운 경우에 활용됩니다.
- 감성 분석이나 텍스트 마이닝 등의 방법을 활용하여 분석하며, 고객 만족도 조사나 제품 리뷰 등에서 활용됩니다.

## 정량적 데이터

- 주로 **양적** 데이터로 불리며, 수치화된 데이터를 의미합니다.  
- 예를 들어, 고개 만족도 조사에서 고객의 만족도를 1점부터 5점까지의 척도로 평가하는 경우가 있습니다.
- 이러한 데이터는 주로 통계적 분석이 가능하며, 인과관계를 분석하거나 예측 모델을 만드는 등의 다양한 분석 작업에 활용됩니다.
- 통계 분석이나 데이터 마이닝 등의 방법을 활용하여 분석하며, 매출 현황, 고객 구매 패턴 등에서 활용됩니다.

# 정성적 데이터와 정량적 데이터의 차이

- 데이터 분석에서 사용하는 방법이 다릅니다.

+ 정량 데이터는 수학적 모델링, 통계 분석 등의 방법을 사용하여 처리 및 분석이 가능합니다.
+ 정성 데이터는 주관적인 요소가 많기 때문에, 주관적인 해석과 판단이 필요하며, 통계 분석 등의 방법이 제한적입니다.

* 정성 데이터는 감성 분석, 이미지 분석 등과 같이 기술적인 부분이 적고, 인간의 주관적인 판단이 큰 역할을 하기 때문에 비교적 해석이 어려운 경우가 많습니다.
* 정량 데이터는 분석 방법이 다양하고, 수학적 모델링을 통해 정확한 결과를 도출할 수 있기 때문에 분석과 해석이 상대적으로 쉽습니다.
