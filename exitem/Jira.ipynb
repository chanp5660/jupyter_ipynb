# Jira

```
요약

지라는 협업에 필요한 흐름을 알 수 있고 누가 어떤 일을 진행하고 있는지 물어보지 않고도 확인할 수 있습니다.
업무의 방향을 확실하게 확인할 수 있어 본인의 업무의 방향에도 도움이 됩니다.

문서로서는 능력은 Confluence를 활용하는게 좋습니다.

```

Jira Software 사용하기[https://www.atlassian.com/ko/software/jira/features](https://www.atlassian.com/ko/software/jira/features)

소프트웨어 개발 및 프로젝트 관리를 위한 **애자일** 기반의 이슈 트래킹 도구입니다. Atlassian이라는 회사에서 개발하고 유지 보수합니다.

**이슈**를 생성하고, 할 일을 추적하고, 업무의 진행 상황을 모니터링하며, 팀 간 협업을 촉진하는 기능을 제공합니다. 이슈는 각각 고유한 키를 가지면, 상태, 우선순위, 할당된 담당자, 직업 진행 상황 등의 정보를 포함합니다.

강력한 대시보드 및 보고 기능도 제공합니다. 프로젝트 상태, 팀의 업무 진행 상황, 직업에 대한 수행 시간 및 예산 등의 정보를 시각적으로 나타내서 보고서를 생성하고 프로젝트 관리를 돕습니다. 

이슈를 관리하는 데에 유용한 여러 가지 기능을 제공합니다. 예를 들어, 이슈에 대한 코멘트를 남기고, 파일을 첨부하고, 이슈를 라벨링하고, 이슈를 추적하기 위한 워크플로우를 정의할 수 있습니다. 이슈를 할당하고, 작업 시간을 추적하며, 이슈에 대한 알림을 설정할 수도 있습니다.

대규모 프로젝트를 위한 확장성과 유연성을 제공합니다. 여러 팀이 Jira를 사용하고 각 팀마다 자신만의 프로젝트, 워크플로우, 이슈 유형 등을 정의할 수 있습니다. 또한 Jira는 다양한 플로그인을 지원하여 사용자 요구 사항에 맞게 기능을 확장할 수 있습니다.

---
또한 소프트웨어 개발 뿐만 아니라 다양한 프로젝트와 업무 관리를 위한 툴로도 사용됩니다. 팀 전체에서 작업의 진행 상황을 실시간으로 모니터링할 수 있으며, 이슈가 **누락**되거나 **중복**되지 않도록 체계적으로 관리할 수 있습니다.

각각의 이슈를 티켓(Ticket)으로 표현합니다. 티켓은 제목, 설명, 우선순위, 담당자, 상태 등의 정보를 담고 있으며, 이슈의 종류에 따라서는 추가적인 필드를 설정할 수 있습니다. 이슈를 생성하면 해당 이슈에 대한 토론과 댓글 작성, 파일 첨부 등의 기능을 활용하여 팀원 간 소통과 협업을 할 수 있습니다.

이슈를 추적하기 위한 워크플로우(Workflow)를 지원합니다. 워크플로우는 이슈가 생성되어 해결될 때 이슈의 상태가 변경되는 일련의 과정을 정의하는 것입니다. 예를 들어, 소프트웨어 개발 프로젝트에서는 워크플로우를 "To do", "In progress", "Done"과 같이 정의하여, 이슈가 생성되면 초기 상태는 "To do"이 되고, 개발이 시작되면 "In progress" 상태로 변경되며, 완료되면 "Done" 상태로 변경됩니다. 이를 통해 팀원들은 이슈가 어느 단계에 있는지 쉽게 파악할 수 있습니다.

대시보드(Dashboard)를 통해 프로젝트의 전반적인 상황을 한눈에 파악할 수 있습니다. 대시보드에는 프로젝트의 진행 상황, 이슈의 우선순위, 팀원의 일정 등의 정보가 표시되며, 개인적으로 자주 사용하는 이슈나 대시보드를 즐겨찾기에 등록하여 효율적으로 이용할 수 있습니다.

또한 다양한 플러그인(Plugin)을 지원하여, 사용자가 필요에 따라 기능을 확장할 수 있습니다. 예를 들어, 코드 리뷰를 위한 플러그인인 Crucible이나 소스 코드 버전 관리를 위한 Bitbucket 등의 툴을 연동하여 더욱 효율적인 개발 환경을 구성할 수 있습니다.

종합하면, 이슈 트래킹과 워크플로우, 대시보드, 플러그인 등 다양한 기능을 제공하여 프로젝트 관리를 보다 체계적으로 할 수 있도록 도와주는 프로젝트 관리 툴입니다. 또한, 사용하면 이슈의 우선순위, 담당자, 상태 등을 쉽게 파악하여 팀원 간의 협업을 원활하게 할 수 있습니다. 더불어, Jira를 통해 프로젝트의 진행 상황을 실시간으로 모니터링할 수 있어, 프로젝트 관리에 필요한 데이터를 신속하게 수집하고 분석할 수 있습니다.

무료로 제공되는 버전과 유료로 제공되는 버전이 있습니다. 무료 버전은 기본적인 이슈 추적과 워크플로우 관리 기능만을 제공하며, 유료 버전에는 추가적으로 프로젝트 관리, 보고서 생성, 통계 분석 등의 기능이 제공됩니다.

많은 기업에서 프로젝트 관리에 활용하고 있으며, 소프트웨어 개발뿐만 아니라 마케팅, 인사 등 다양한 부서에서도 사용되고 있습니다. Jira의 다양한 기능과 확장성은 프로젝트 관리에 있어 매우 유용한 도구 중 하나입니다.
