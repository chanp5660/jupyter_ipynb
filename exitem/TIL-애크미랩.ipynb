> 여러 방면으로 경험이 있으신 경력자가 뽑혀야 할 것 같다.  

>이제 3년차가 되는 스타트업으로 사원수가 8명인 정도에 벤처 기업으로 투자를 받아 회사가 운영되는 것은 일반적인 루트인 것 같다. 아직 정보가 적지만 찾아본 내용으로만 생각해보면 여러 트랜드를 알 수록 회사에 도움이 될 것 같고 파이썬 프로그램 개발자가 다방면(백엔드, 프론트)으로 활동을 해야할 것 같은 느낌이다. 아직 체계가 잡히지 않아 당연하지만 신입으로서는 능력자 사수가 있지 않은 이상 꽤 어려운 길일 것 같다는 생각이 들었다. 물론 내가 만들어가는 것도 좋지만 나의 도전들이 누군가에게 피해를 줄 수 있다면 좋은 방법이 아니라고 본다. 경력자로서 여러 방면으로 활동해보신 분들이 새롭게 자신의 길을 이끌어 가기에는 투자도 있고 좋은 기회라고 생각한다.

## 구체적인 사전 테스트
[해당 사이트](https://www.acmelab.co.kr/783dd19e-f629-4c87-8a73-244bdb2268d7)

1. 텍스트 마이닝을 활용한 뉴스레터  

    - 배경 및 니즈
        - 회사에는 컨텐츠 제작자, 마케터, 엠디 등 다양한 직군이 있고, 당사가 취급하는 자사 제품(화장품) 뿐만 아니라 브랜드 파트너(건강식품, 농산물 등)도 가지고 있으며, 이를 라이브 커머스 뿐만 아니라 이커머스 플랫폼을 활용해서 판매 활동을 진행함.
        - 즉, 다양한 직군의 사람들이 다양한 영역에 대해서 관심를 가지고 있으며, 각 트랜드를 명확하게 파악하고 실행하는 것이 개인의 성과에 직접적인 영향을 준다고 볼 수 있음.
        - 이를 위한 뉴스레터를 텍스트 마이닝을 활용해서 주기적(매일) 회사 구성원들과 파트너에게 공급하고 싶습니다.
        - 이를 위한 파이썬 프로그램을 해주세요.

    - 제약조건
        - 주요 키워드는 회사의 구성원들이 넣을 수 있어야 함.
        - 회사 구성원들을 포함한 구독자들의 구독율이 해당 정보에 대한 피드백으로 인지되었으면 하고, 이를 반영해서 구독율이 높은 영역의 뉴스가 더 많이, 더 깊게 다뤄졌으면 하며, 그렇지 않은 주제는 많이 다뤄질 필요는 없어 보임.
        - 한국어의 특성에 따라 해당 키워드가 있는 뉴스를 클리핑 해온다는게 무의미한 경우가 있음. 이를 해결할 수 있는 방안이 필요함.
        - 뉴스레터를 읽기 편한 모델로 만들어서 배포 해주었으면 함.
        - 퍼블리싱은 노션을 통한 퍼블리싱이었으면 함. 그외 이메일 및 슬랙 notification이 되었으면 함.
    
2. 인스타그램기반의 가상 인격체(vitual identity)를 성장시킬 수 있는 프로그램 개발
    - 배경
        - 인스타그램을 통해서 자사가 관리해야하는 자사 공식 페이지 뿐만 아니라 파트너 페이지가 존재함.
        - 또한 자사는 몇개의 가상의 아이디를 가지고 있음. 이 아이디를 활용해서 자사가 관리하는 자사 및 파트너 인스타그램을 활성화 시키고자 함.
            - 활성화의 정의 : 해당 인스타그램의 좋아요 증가, 댓글 및 공유 증가, 게시물 증가
        - 각 아이디는 봇이라는 이미지를 주지 않겠금 하기 위한 다양한 노력이 필요함. 이 노력의 결과로 각 아이디는 특정 아이덴터티를 부여하고 그에 맞는 인스타그램 활동을 진행했으면 함.
        - 이를 진행할 수 있는 파이썬 프로그래밍을 해주세요
    - 제약조건
        - 자동로그인 및 대상 페이지에 좋아요를 하는 것은 기본이나, 각 아이디마다의 적절한 인스타그램 활동(게시, 타 페이지의 좋아요, 댓글, 공유 활동)을 하지 않으면 봇으로 간주되어 우리가 키우려고 하는 버츄얼 아이덴터티와는 거리감이 있음
        - 사람이 만들어놓은 게시물을 게시하는 것보다는 기본 이미지, 동영상 repository를 활용해서 게시물을 자동 편집, 게시 할 수 있어야함.
